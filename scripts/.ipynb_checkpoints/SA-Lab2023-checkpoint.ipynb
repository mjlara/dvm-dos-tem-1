

The `SensitivtyDriver` object is designed to help setup and conduct a sensitivity analysis. There are some hard-coded assumptions in the driver object (which pixel to run, the input dataset/site to use, the source of the initial parameter values, which outputs to process and the location of the working directory). The driver object is configurable with respect to the parameters to be used in the sensitivity analysis, the bounds for the parameters, which PFT to analyze, and which CMT (community type) to run.

Let's design our experiment. We need assign number of smaples, select desired parameters, choose percent variance from the initial value, and sampling method. Currently we have two sampling methods available to us: latin hypercube ([lhs](https://en.wikipedia.org/wiki/Latin_hypercube_sampling)) and [uniform sampling](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html).

### 3. SA output visualization

```python

def get_samples(fdir,oname,TIME):
    #make a list of all output files  
    
    fname = oname+'_'+TIME+'_tr.nc'
    fsamples = []
    for item in os.listdir(fdir):
        if item[:8]=='sample_0':
            fsamples.append(fdir+item+'/output/'+fname)
    fsamples.sort()

    return fsamples

def get_output_var_size(fdir,oname,TIME):   
    filename = fdir+item+'/output/'+oname+'_'+TIME+'_tr.nc'
    stg_data = nc.Dataset(filename, 'r')
    print(oname,'dim:',len(stg_data.dimensions))

    
def netcdf3D_df(fdir,oname,TIME):
    '''
    Input:
        fdir: path to the sensitivity analysis directory (workdir)
        oname: output name
    Output:
        df1: dataframe of oname time series for tr run
    '''    
    fsamples = get_samples(fdir,oname,TIME)
    #get the 'oname' variable time series from the correspoding output nc files  
    stg_data = nc.Dataset(fsamples[0], 'r')
    if TIME == 'monthly' :
      #equals how many years of monthly data to process (i.e period=12, prints the last year). If this value is changed then period_range should be change accordingly as well
      period = 1380 
    elif TIME == 'yearly':
      #equals how years of data to process. If this value is changed then period_range should be change accordingly as well
      period = 115 
    if TIME == 'monthly' :
      freq = 'M'
    elif TIME == 'yearly':
      freq = 'Y'   
    
    df1 = pd.DataFrame(stg_data.variables[oname][-period:,0,0], columns=[oname+'_0'])   
    for i in range(1,len(fsamples)):
        stg_data = nc.Dataset(fsamples[i], 'r')
        df2 = pd.DataFrame(stg_data.variables[oname][-period:,0,0], columns=[oname+'_'+str(i)])
        df1 = df1.join(df2)
        #if number of periods has been modified, period_range should be changed accordingly
    df1 = df1.set_index(pd.Series(pd.period_range("1/1/1901", freq=freq, periods=period)))
    sm_fname = driver.work_dir+'/'+oname+".csv"
    df1.to_csv(sm_fname)
    return df1


def netcdf4D_orig(fdir,oname,TIME,idx):
    '''
    Input:
        fdir: path to the sensitivity analysis directory (workdir)
        oname: output name
        idx: layer or pft number
    Output:
        df1: dataframe of oname time series for tr run
    '''

    fsamples = get_samples(fdir,oname,TIME)
    
    #get the 'oname' variable time series from the correspoding output nc files  
    stg_data = nc.Dataset(fsamples[0], 'r')

    if TIME == 'monthly' :
      period = 1380
    elif TIME == 'yearly':
      period = 115
    if TIME == 'monthly' :
      freq = 'M'
    elif TIME == 'yearly':
      freq = 'Y'   
      
    df1 = pd.DataFrame(stg_data.variables[oname][-period:,idx,0,0], columns=[oname+'_0'])   
    for i in range(1,len(fsamples)):
        stg_data = nc.Dataset(fsamples[i], 'r')
        df2 = pd.DataFrame(stg_data.variables[oname][-period:,idx,0,0], columns=[oname+'_'+str(i)])
        df1 = df1.join(df2)
    df1 = df1.set_index(pd.Series(pd.period_range("1/1/1901", freq=freq, periods=period)))
    sm_fname = driver.work_dir+'/'+oname+".csv"
    df1.to_csv(sm_fname)
    return df1
#Call your directory 
fdir=driver.work_dir+'/'
# Specify the outputs you want to visualize

ald = netcdf3D_df(fdir,'ALD','yearly')
gpp = netcdf3D_df(fdir,'GPP','monthly')
snowdepth = netcdf3D_df(fdir,'SNOWTHICK','monthly')
```

# 3. SA Visualize Data-Model Comparison
###  Compute Root Mean Square Error (RMSE)

```python
homedir='/data/workflows/'
def filter_by_date(VAROUT,path_to_obser,path_to_model):
  '''
    VAROUT: variable name output
    path_to_obser: path to observations
    path_to_model: path to post-processed modeled data
  '''
  # NOTE: for soil temperature (tlayer) specify the depth as well for VAR='10cm_tlayer4", '20cm_tlayer5', '40cm_tlayer7'
  obs=pd.read_csv(path_to_obser)
  out=pd.read_csv(path_to_model)
  print(VAROUT+' observ shape (before):',obs.shape)
  print(VAROUT+' TEM outputs shape (before):',out.shape)

  obs.dropna(inplace=True)
  obs=obs.set_index('date')
  formatted_vars=['GPP','SNOWTHICK','ALD']
  if VAROUT in formatted_vars:
    obs.index = obs.index.astype('datetime64[ns]')  #Comment this out for yearly observations (ald, deepdz, shlwdz)
  
  out=out.rename(columns={'Unnamed: 0':"date"})
  out=out.set_index('date')
  out.index=out.index.astype('datetime64[ns]') 
  common_index = set(obs.index).intersection(out.index)
  obs = obs.loc[common_index].sort_values(by='date')
  out = out.loc[common_index].sort_values(by='date')
  print(VAROUT+' observ shape (after):',obs.shape)
  print(VAROUT+' TEM outputs shape (after):',out.shape)
  print()
  return [obs,out]

def plot_obs_vs_mod(VARNAME,obs,out): 
  '''
  VARNAME: variable name
  obs: observation data 
  out: modeled data
  '''
  plt.figure(figsize=(10,4))
  plt.plot(obs,'r',linewidth=2)
  plt.plot(out,'b',alpha=0.5);

  plt.xticks(rotation=90)
  plt.ylabel(VARNAME);plt.xlabel('Time')
  plt.legend(['observed','simulated']);


def rmse_SA(VAROBS,VAROUT,fplot=False):
  '''
  VAROBS: variable name observaion
  VAROUT: variable name output
  '''
  print('compiling data for data for:', VAROBS)
  path_to_obser="/data/input-catalog/sensitivity_analysis_lab/EML/ak_"+VAROBS+".csv"
  # path_to_obser=fdir+"EML/observed/ak_"+VAROBS+".csv"

  path_to_model=driver.work_dir+'/'+VAROUT+".csv"
  # path_to_model=fdir+"EML/modeled/EML1000_"+VAROUT+".csv"

  [obs,out]=filter_by_date(VAROUT,path_to_obser,path_to_model)
  if fplot:
    plot_obs_vs_mod(VAROBS,obs,out)

  return np.sqrt((out.sub(obs['value'], axis=0)**2).mean())

X_sample=pd.read_csv(fdir+"sample_matrix.csv")


### MODIFY THIS TO YOUR VARIABLES, currently defaulted to GPP, ALD, SNOWTHICK ###
rmse_gpp=rmse_SA('eml_cipehr_gpp','GPP',1)
rmse_alt=rmse_SA('eml_cipehr_ald','ALD',1)
rmse_snowthick=rmse_SA('eml_cipehr_snowdepth','SNOWTHICK',1)

```

## 4. Calculating the parameter importance using Random Forest

```python


def find_important_features(X,y,fplot=False,fname=''):
  model = RandomForestRegressor()

  X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                      test_size=1, random_state=0)

  model.fit(X_train, y_train)
  y_pred=model.predict(X_test)

  print(fname + f' model score on data: {model.score(X_train, y_train)}')

  #import matplotlib.pyplot as plt
  if fplot:

    importances = model.feature_importances_
    indices = np.argsort(importances)

    fig, ax = plt.subplots(figsize=(5, 10))
    ax.barh(range(len(importances)), importances[indices])
    ax.set_yticks(range(len(importances)))
    d = ax.set_yticklabels(np.array(X_train.columns)[indices])
    plt.title(fname)
    
    top_3_names =np.array(X_train.columns)[indices][-3:]
    fig,ax=plt.subplots(1,len(top_3_names),figsize=(10, 5))
    for i,x in enumerate(top_3_names):
        slope, intercept, _, _, _ = linregress(X_sample[x].values,y.values)
        ax[i].plot(X_sample[x].values,y.values,'.'); ax[i].set(xlabel=x, ylabel='RMSE');
        ax[i].plot(X_sample[x].values, slope * X_sample[x].values + intercept, 'r')

    plt.tight_layout()    

  return model
###EML###
gppmodel=find_important_features(X_sample,rmse_gpp,1,'GPP')
altmodel=find_important_features(X_sample,rmse_alt,1,'ALT')
snowthickmodel=find_important_features(X_sample,rmse_snowthick,1,'SNOWTHICK')
```
